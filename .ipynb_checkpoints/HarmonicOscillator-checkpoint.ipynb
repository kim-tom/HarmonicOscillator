# 陽的Euler法 #

```julia
using PyPlot
```

```julia
dt = 0.05
function explicit_euler(p,q)
    p_new = p-q*dt
    q_new = q+p*dt
    return p_new, q_new
end
p0, q0 = 1.0, 0.0
p, q = p0, q0
pq = [p, q]
num =Int(div(4*pi,dt))
for i in 1:num
    p,q = explicit_euler(p,q)
    pq = hcat(pq, [p,q])
end
xlabel("p")
ylabel("q")
grid("on")
axis("equal")
plot(pq[1,1:num+1],pq[2,1:num+1])
```

# 陰的Euler法 #

```julia
dt=0.05
function implicit_euler(p,q)
    denom=1+dt^2
    p_new = (p-dt*q)/denom
    #qは除算無しで計算できる。
    q_new = dt*p_new+q
    return p_new, q_new
end
p0, q0 = 1.0, 0.0
p, q = p0, q0
pq = [p, q]
num =Int(div(4*pi,dt))
for i in 1:num
    p,q = implicit_euler(p,q)
    pq = hcat(pq, [p,q])
end
xlabel("p")
ylabel("q")
grid("on")
axis("equal")
plot(pq[1,1:num+1],pq[2,1:num+1])
```

# 台形公式

```julia
dt=5
function trapezoidal(p,q)
    dt2=dt*dt
    denom=4+dt^2
    p_new = ((4-dt2)*p-4dt*q)/denom
    #qは除算無しで計算できる。
    q_new = dt*(p_new+p)/2+q
    return p_new, q_new
end
p0, q0 = 1.0, 0.0
p, q = p0, q0
pq = [p, q]
num =Int(div(50*pi,dt))
for i in 1:num
    p,q = trapezoidal(p,q)
    pq = hcat(pq, [p,q])
end

xlabel("p")
ylabel("q")
grid("on")
axis("equal")
plot(pq[1,1:num+1],pq[2,1:num+1])
```

# Runge-Kutta法
    1.安定な例

```julia
dt=0.05
function RungeKutta(p,q)
    A=(24-12*dt^2+dt^4)/24
    B=dt*(dt^2-6)/6
    p_new = A*p+B*q
    q_new = A*q-B*p
    return p_new, q_new
end
p0, q0 = 1.0, 0.0
p, q = p0, q0
pq = [p, q]
num =Int(div(2pi,dt))
for i in 1:num
    p,q = RungeKutta(p,q)
    pq = hcat(pq, [p,q])
end

xlabel("p")
ylabel("q")
grid("on")
axis("equal")
plot(pq[1,1:num+1],pq[2,1:num+1])
```

    2.不安定な例

```julia
dt=2.9
function RungeKutta(p,q)
    A=(24-12*dt^2+dt^4)/24
    B=dt*(dt^2-6)/6
    p_new = A*p+B*q
    q_new = A*q-B*p
    return p_new, q_new
end
p0, q0 = 1.0, 0.0
p, q = p0, q0
pq = [p, q]
num =Int(div(20pi,dt))
for i in 1:num
    p,q = RungeKutta(p,q)
    pq = hcat(pq, [p,q])
end
xlabel("p")
ylabel("q")
grid("on")
axis("equal")
plot(pq[1,1:num+1],pq[2,1:num+1])
```
